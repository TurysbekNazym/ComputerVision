### Task 3
Shake the image with a noise like "salt and pepper". Suppress noise with a median filter. Map the difference between the original and noisy images, and between the adjusted and the original.

```python
#median filter
import matplotlib.pyplot as plt
import matplotlib.image as img
import scipy.ndimage as ndimage
%matplotlib inline
im = img.imread('./ove.jpg')
fig = plt.figure(figsize=(15,10))
plt.subplot(131)
im_med = ndimage.median_filter(im, 3)
plt.title('Median filter', fontsize=20)
plt.imshow(im_med,cmap='gray' )
plt.subplot(132)
plt.title('Original', fontsize=20)
plt.imshow(im,cmap='gray')
plt.subplot(133)
plt.title('Result', fontsize=20)
plt.imshow((im-im_med),cmap='gray')

plt.show()
```

```python
#median filter original - result
import matplotlib.pyplot as plt
import matplotlib.image as img
import scipy.ndimage as ndimage
%matplotlib inline
im = img.imread('./ove.jpg')
fig = plt.figure(figsize=(15,10))
plt.subplot(131)
im_med = ndimage.median_filter(im, 3)
plt.title('Median filter', fontsize=20)
plt.imshow(im_med,cmap='gray' )
plt.subplot(132)
plt.title('Original', fontsize=20)
plt.imshow(im,cmap='gray')

plt.subplot(133)
plt.title('original - result', fontsize=20)
plt.imshow(((im-im_med)-im),cmap='gray')
plt.show()
```
