```python
#grayscale
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
fig = plt.figure(figsize=(15,10))
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])

img = mpimg.imread('person.jpg')     
gray = rgb2gray(img)    


plt.subplot(221)
plt.title('result', fontsize=20)
plt.imshow(gray,plt.get_cmap('gray'))

plt.subplot(222)
plt.title('original', fontsize=20)
plt.imshow(img,cmap='gray')
plt.show()
```

```python
##linear stretching 
import matplotlib.pyplot as plt
import matplotlib.image as img
from scipy import ndimage as nimg
import numpy as np
from skimage import color

image = img.imread('person.jpg')
fig = plt.figure(figsize=(25,20))
imbook = color.rgb2gray(image)
gflt = nimg.gaussian_filter(imbook, sigma=20)
fig = plt.figure(figsize=(15,10))
plt.subplot(132)
plt.imshow(imbook,cmap='gray')
plt.imshow((imbook-gflt),cmap='gray')
maxValue = (imbook).max();
minValue = (imbook).min();
newImage = (imbook-minValue)* ((255-0)/(maxValue-minValue))
plt.subplot(133)
plt.imshow(newImage,cmap='gray')
plt.show()
 
```

```python
# importing required libraries of opencv
import cv2
 
# importing library for plotting
from matplotlib import pyplot as plt
 
# reads an input image
img = cv2.imread('person.jpg',0)
 
# find frequency of pixels in range 0-255
histr = cv2.calcHist([img],[0],None,[255],[0,255])
 
# show the plotting graph of an image
plt.plot(histr)
plt.show()
```
